```python
#Exercise 3: Find the most expensive car company name
import pandas as pd
df=pd.read_csv("C:\\Users\\Apeksha\\Downloads\\Automobile_data (1).csv")
df["prices"]=pd.to_numeric(df["price"],errors="coerce")# tryied to convert the value in prices to numerical  like (int flot)
#errors="coerce") means:"If there's a problem converting, don't crash â€“ just use NaN."
df = df.dropna(subset=['price']) #This removes any rows where the price column is NaN (missing).
max_prices=df["prices"].max() #This finds the maximum numeric value in the price column.
result=df[df["prices"]==max_prices][['company','price']]
print(result)
```

```python
#Exercise 9: Concatenate two data frames using the following conditions
import pandas as pd

Germancars={"Company":['Ford','Mercedes','Bmv','Audi'],
            "Price":[23845,171995,135925,71400]}
Carsdf1=pd.DataFrame.from_dict(Germancars)

japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 
                'Price': [29995, 23600, 61500 , 58900]}
carsdf2 = pd.DataFrame.from_dict(japaneseCars)

Carsdf=pd.concat([Carsdf1,carsdf2],keys=["Germany","Japan"])

print(Carsdf)
```

```python
#Exercise 10: Merge two data frames using the following condition
import pandas as pd
Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
Car_PriceDf=pd.DataFrame.from_dict(Car_Price)

car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}
car_HorsepowerDf=pd.DataFrame.from_dict(car_Horsepower)

CarsDf=pd.merge(Car_PriceDf,car_HorsepowerDf,on="Company")
#CarsDf
print(CarsDf)
```
